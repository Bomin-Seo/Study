3) 주어진 for 문을, 코드 한 줄만 사용하여 출력화면과 같이 나오도록 x2, x3, x4 리스트를 만들도 록 코드를 작성하시오.

```python
x2 = [i*i for i in range(10)]
print(x2)

x3=[[i,j*2] for i in [10,2,3,1] for j in [2,4]]
print(x3)

x4=[[i,j*2] for i in [10,2,3,1] for j in [2,4] if i!=j]
print(x4)
```

3. matplotlib를 사용한 차트 그리기 1) 다음의 x,y 리스트가 주어졌을 때 출력화면과 같이 나오도록 실선 그래프와 막대 그래프르 matplotlib으로 그리는 코드를 작성하시오. (matplotlib만 사용)

```python
import matplotlib.pyplot as plt
import numpy as np
x = [1,2,3,4,5,6,7]
y = [12,20,25,22,20,25,30]

plt.plot(x,y)
plt.bar(x,y)
plt.show()
```

2) 다음의 x,y 리스트가 주어졌을 때 선의 종류와 모양, legend, label, title을 사용하여 출력화면과 같이 나오도록 그래프를 꾸미는 코드를 작성하시오. (matplotlib만 사용)

```python
x = np.array([1,2,3,4,5,6,7])
y = np.array([[12,14,40],[20,25,43],[25,30,41],[22,20,35],[20,30,42],[25,30,35],[30,35,38]])

plt.title('Amount of sales')
plt.xlabel('days')
plt.ylabel('amount(unit: Kg)')
plt.plot(x, y[:,0], marker = '*', color = 'green', linewidth=1.5,markersize=8)
plt.plot(x, y[:,1], marker = 'o', color = 'red', linestyle = '--', linewidth=1.5,markersize=8)
plt.plot(x, y[:,2], marker = '+', color = 'blue', linestyle = 'dotted', linewidth=1.5,markersize=8)
plt.legend(['apple','plum','strawberry'], loc = 4)
plt.show()
```

3) 다음의 x,y 리스트가 주어졌을 때 출력화면과 같이 나오도록 subplot을 사용하여 여러 그래프 를 하나의 화면에 그리는 코드를 작성하시오. (matplotlib만 사용)

```python
x = np.array([1,2,3,4,5,6,7])
y = np.array([[12,14,40],[20,25,43],[25,30,41],[22,20,35],[20,30,42],[25,30,35],[30,35,38]])

plt.figure(figsize=(15,3))
plt.suptitle("Amount of sales for 3 items", fontsize=20)

ax1 = plt.subplot(1,3,1)
ax1.set_xlabel('days')
ax1.set_ylabel('amount (unit: Kg)')
plt.plot(x,y[:,0], marker = '*', color = 'green',linewidth=1.5,markersize=8)
plt.legend(['apple'], loc = 2)

ax2 = plt.subplot(1,3,2)
ax2.set_xlabel('days')
ax2.set_ylabel('amount (unit: Kg)')
plt.plot(x,y[:,1], marker = 'o', color = 'red', linestyle = '--',linewidth=1.5,markersize=8)
plt.legend(['plum'], loc = 2)

ax3 = plt.subplot(1,3,3)
ax3.set_xlabel('days')
ax3.set_ylabel('amount (unit: Kg)')
plt.plot(x,y[:,2], marker = '+', color = 'blue', linestyle = 'dotted',linewidth=1.5,markersize=8)
plt.legend(['strawberry'], loc = 1)


plt.show()
```
